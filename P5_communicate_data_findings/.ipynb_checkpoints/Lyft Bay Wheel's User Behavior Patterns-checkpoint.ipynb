## Investigation Overview

> We are going to investigate about the user behavior patterns at the lyft bay wheel! The main focus is `time` and `space` related to this bike service. We are going to dive in when and how long user ride the bike and what station is most popular and much more!

## Dataset Overview

> Lyft operate bike sharing service around bay area and collected the data related to this bike sharing service. This dataset is consisted with the 6,017,920 trip data about location, time, and which bike etc opened to the public. Every trip data is anonymized.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
# Load the dataset
df = pd.read_pickle('lyft_bay_wheel_master.pkl')
```

## What time user mostly start to ride a bike?

Users mostly use bike 8-9 am & 5-6 pm. We could assume that maybe the user of bike service are mostly an office worker who commutes by bike. Or this could be a people who is traveling! Now let's look at how long user ride a bike!

![when](image/what_time_user_mostly_start_to_ride_a_bike.png)

```python
# choose the base color
base_color = sb.color_palette()[0]

# Choose the figure size
plt.figure(figsize=[10, 7])

### START TIME ###

# Count how many data exist for each start hour
start_hour = df.start_time.dt.hour
start_hour_dic = start_hour.value_counts()

# make series to dataframe using reset_index()
sb.countplot(data=start_hour.reset_index(), y='start_time', color=base_color)
plt.xlabel('Count', fontsize=12)
plt.ylabel('start time - (hour)', fontsize=12)

# To show the count value of each 
locs, labels = plt.yticks()
for loc, label in zip(locs, labels):
    count = start_hour_dic[int(label.get_text())]
    string = str(count)
    plt.text(count+10000, loc, string, va='center')
    
# Add more ticks to put the annotation inside the figure
new_loc = plt.xticks()[0].astype(int)
plt.title('What time user mostly start to ride a bike?', fontsize=16)
plt.xticks(new_loc, new_loc);
```

## How long user ride the bike?

Distribution is left-skewed and unimodal. In the log scale if we only consider duration second between 50 and 6000 sec we can see that the graph is showing normal distribution and most of the people used bike for 370 second, approximately 6 minutes! Isn't quite interesting that user ride so little? Now we now when and how long user ride a bike! Then where does user ride the bike?

![howlong](image/how_long_user_ride_the_bike.png)

```python
duration_sec = df.duration_sec[~df.duration_sec.isna()]
duration_sec[duration_sec > 100000]

loc = [100, 300, 1000, 3000]
interval = 10
bins = np.arange(50, 6000 + interval, interval)

plt.figure(figsize=[7, 5])
duration_sec.hist(bins=bins)
plt.xscale('log')
plt.xticks(loc, loc)

plt.title('How long user ride the bike?')
plt.xlabel('Riding time (sec)')
plt.ylabel('Count');
```

## Where do user rental there bike and go to mostly (TOP 5)?

For rental, `Market St at 10th St` is the most popular station. For destination, `San Francisco Caltrain (Townsend St at 4th St)` is the most popular station. We know that 8am and 5pm is the most popular usage time and now we know the most popular station. So what station people are using at 8am and 5pm? Will it be the same station that came out here?

![top10](image/where_do_user_rental_there_bike_and_go_to_mostly.png)

```python
start_top10 = list(df.start_station_name.value_counts()[:5].index)
end_top10 = list(df.end_station_name.value_counts()[:5].index)

# choose the base color
base_color = sb.color_palette()[0]

# Choose the figure size
plt.figure(figsize=[10, 5])

### START STATION ###
plt.subplot(211)

# Get only top 10 start station
start_df = df.query('start_station_name in @start_top10').start_station_name

# make series to dataframe using reset_index()
sb.countplot(data=start_df.reset_index(), y='start_station_name', color=base_color,
             order=start_top10)
plt.xlabel('Count', fontsize=12)
plt.ylabel('station name', fontsize=12)

# To show the count value of each 
locs, labels = plt.yticks()
for loc, label in zip(locs, labels):
    # label is not int and we get station name as a string
    # We use this name to get the index from the start_top10
    # which is list containing the station names as decreasing order
    # comparing the total count of usage of each station
    # And this index will be used to get the count number for each station
    count = start_df.value_counts().iloc[start_top10.index(label.get_text())]
    string = str(count)
    plt.text(count-8000, loc, string, va='center', color='w')
    
plt.title('Where do user rental there bike and go to mostly (TOP 5)?', fontsize=16)

### END STATION ###
plt.subplot(212)

# Get only top 10 end station
end_df = df.query('end_station_name in @end_top10').end_station_name

sb.countplot(data=end_df.reset_index(), y='end_station_name', color=base_color,
             order=end_top10)
plt.xlabel('Count', fontsize=12)
plt.ylabel('station name', fontsize=12)

# To show the count value of each 
locs, labels = plt.yticks()
for loc, label in zip(locs, labels):
    count = end_df.value_counts().iloc[end_top10.index(label.get_text())]
    string = str(count)
    plt.text(count-10000, loc, string, va='center', color='w');
```

## What bike station people use most at 8am and 5pm?

`San Francisco Caltrain` is the station most popular at **8am and 5pm**. There is a difference of amount of usage to or from **public transportation** based on the time but still `8am` and `5pm` has the most usage for both `start station` and `end station` near **public transportation**. We looked at the stations enough. Additionally we could get curious where does service is available? Let's dive in to this!

![58](image/what_bike_station_people_user_most_at_8_5.png)

```python
bike_df = df.copy()

# Generate column "hour"
bike_df['hour'] = bike_df.start_time.dt.hour

# leave the necessary columns
bike_df = bike_df[['hour', 'start_station_name', 'end_station_name']]

# Remove NaN values
bike_df = bike_df[~(bike_df.hour.isna() |
                    bike_df.start_station_name.isna() |
                    bike_df.end_station_name.isna())]

# Leave only 8am and 5pm
bike_df_8_17 = bike_df.query('hour in [8, 17]')

top5 = list(bike_df_8_17.start_station_name.value_counts().iloc[:5].index)

plt.figure(figsize=[7, 7])

# violinplot
plt.subplot(211)
plt.title('What bike station people use most at 8am and 5pm?', fontsize=16)

sb.violinplot(data=bike_df.query('start_station_name in @top5'),
              x='hour', y='start_station_name', inner=None)
plt.xlabel('start hour')
plt.ylabel('station name')

# boxplot
plt.subplot(212)
sb.boxplot(data=bike_df.query('start_station_name in @top5'),
           x='hour', y='start_station_name')
plt.xlabel('start hour')
plt.ylabel('station name');
```

## Where is lyft bay wheel service is available?

**Lyft bike station available** at San Francisco, East Bay, San Jose. We can see it from longitude and latitude! But this is hard to look at! So let's use image with map!

![latlng](image/where_is_lyft_service_available_lat_lng.png)

```python
# check the start station lat lng range
start_station_lat_range = df.start_station_latitude.min(), df.start_station_latitude.max()
start_station_lng_range = df.start_station_longitude.min(), df.start_station_longitude.max()

# check the end station lat lng range
end_station_lat_range = df.end_station_latitude.min(), df.end_station_latitude.max()
end_station_lng_range = df.end_station_longitude.min(), df.end_station_longitude.max()

start_lng = df.start_station_longitude[~df.start_station_longitude.isna()]
start_lat = df.start_station_latitude[~df.start_station_latitude.isna()]

end_lng = df.end_station_longitude[~df.end_station_longitude.isna()]
end_lat = df.end_station_latitude[~df.end_station_latitude.isna()]

# Import patches for Rectangular drawing
import matplotlib.patches as patches
import urllib

fig = plt.figure(figsize=[15, 7])
fig.subplots_adjust(top=0.92)
fig.suptitle('Where is lyft bay wheel service is available?', fontsize=20)

interval = 0.01

# start station longitude
plt.subplot(221)

bins = np.arange(-122.5, -121.8 + interval, interval)
start_lng.hist(bins=bins)

plt.xlim(-122.5, -121.8 + interval)
plt.xlabel('Start station Longitude')
plt.ylabel('Count')

# start station latitude
plt.subplot(222)

bins = np.arange(37.25, 37.9 + interval, interval)
start_lat.hist(bins=bins)

plt.xlim(37.25, 37.9 + interval)
plt.xlabel('Start station Latitude')
plt.ylabel('Count')

# end station longitude
plt.subplot(223)

bins = np.arange(-122.5, -121.8 + interval, interval)
end_lng.hist(bins=bins)
plt.xlim(-122.5, -121.8 + interval)
plt.xlabel('End station Longitude')
plt.ylabel('Count')

# end station latitude
plt.subplot(224)

bins = np.arange(37.25, 37.9 + interval, interval)
end_lat.hist(bins=bins)

plt.xlim(37.25, 37.9 + interval)
plt.xlabel('End station Latitude')
plt.ylabel('Count');
```

Now we can see clear that this service is held inside bay area! So let's look at how the user move around in this area!

![service](image/where_is_lyft_service_available.png)

```python
fig = plt.figure(figsize=[15, 7])

# Add subplot for drawing rectangular
ax = fig.add_subplot(111)

width = start_station_lng_range[1] - start_station_lng_range[0]
height = start_station_lat_range[1] - start_station_lat_range[0]
ax.add_patch(patches.Rectangle((start_station_lng_range[0], start_station_lat_range[0]), width, height, alpha=0.5))

plt.xlim((-125, -120))
plt.ylim((36, 38))

# create an file object from the url
f = urllib.request.urlopen('https://i.pinimg.com/originals/52/bc/83/52bc835a15d22daeadf44fd316dd2a25.jpg')

# Read the file object and convert to np.array
image = plt.imread(f, format='gif')

# Manual xlim ylim tuning
ax.imshow(image, extent=[-125, -113.8, 32.3, 42.4])

plt.xlim((-125, -120))
plt.ylim((36, 38))

plt.title('Where is lyft bay wheel service is available?', fontsize=16)
plt.xlabel('Longitude', fontsize=14)
plt.ylabel('Latitude', fontsize=14);
```

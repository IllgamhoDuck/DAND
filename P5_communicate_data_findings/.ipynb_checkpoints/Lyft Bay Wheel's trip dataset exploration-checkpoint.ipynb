```python
# Get the html data using selenium
html = driver.page_source

# wrap the html data with beautifulsoup
soup = BeautifulSoup(html, 'html.parser')

# Get way to get the url using the regex expression
csv_list = soup.find_all(text=re.compile('csv.zip'))
```

We are going to save the files to `data` folder.

```python
# generate new folder name "data" if there is no folder name "data" exist
dataset_path = 'data'
if not os.path.isdir(dataset_path):
    os.mkdir(dataset_path)

for csv in csv_list:
    # Use the filename as it was indicated in the homepage
    url = csv.parent.attrs['href']
    filename = url.split('/')[-1]
    
    # Request the data programmatically
    try:
        response = requests.get(url)
        with open(os.path.join(dataset_path, filename), mode='wb') as f:
             f.write(response.content)
    except requests.exceptions.RequestException as e:
        raise SystemExit(e)
```

### 3. Unzip the dataset and merge to one csv file

The files are all consisted with zip files so we need to unzip it and merge it to one dataset

```python
import zipfile
import io
```

```python
dataset_path = 'data'

# check the dataset folder exist
if not os.path.isdir(dataset_path):
    raise FileNotFoundError("There is no dataset folder called 'data'")

# We are not going to extract the zip file
# Rather we are just going to open one zip file by one and read the content
# and wrap it with pandas DataFrame to make one big merged csv file
# includes every content of all the zip files

# The list where we are going to store all the csv files
all_csv_list = []

for data in os.listdir(dataset_path):
    # zip file path
    source_file = os.path.join(dataset_path, data)
    
    # Wrap the zip file with Zipfile library
    with zipfile.ZipFile(source_file, 'r') as zf:
        
        # Open the content without extracting the zip file
        with zf.open('.'.join(data.split('.')[:-1])) as f:
            
            # Read the content. This is binary datatype
            binary = f.read()
            
            # Decode the binary file by utf-8. This will return string
            utf8_str = binary.decode('utf-8')
            
            # To use pandas read_csv change string to string file using io.StringIO
            str_file = io.StringIO(utf8_str)
            
            # Wrap the content with pandas DataFrame temporarily.
            # We are going to store the csv file to one master csv file
            tmp_csv = pd.read_csv(str_file)
            
            # Append to the list
            all_csv_list.append(tmp_csv)
```

```python
# Check the csv list for a while
all_csv_list[2].head(5)
```

```python
# Concat all the csv files to one master csv file
# ignore_index ignore the original indebx 
master_csv = pd.concat(all_csv_list, axis=0, ignore_index=True)

# save it to lyft_bay_wheel.csv file
master_csv.to_csv("lytf_bay_wheel.csv", index=False)
```

### 3. Tidiness issues

[Tidy data](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html) is a standard way of mapping a dataset to its structure. Dataset is messy or tidy depending on how rows, columns and tables are matched up with observations, variables and types.

**Tidy data follows 3 rules**

1. Each variable forms a column
2. Each observation forms a row
3. Each type of observational unit forms a table

**5 most common tidiness issues**

1. Columns headers are values, not variable names
2. Multiple variables are stored in one column
3. Variables are stored in both rows and columns
4. Multiple types of observational units are stored in the same table
5. A single observational unit is stored in multiple tables

### 4. Quality issues

Data quality is related to the content. If there is issues with content we say dirty data. It could be inaccurate, corrupted, or duplicated data. We can consider **Data quality dimensions** when we deal with quality issues.

**Data quality dimensions**

- **Completeness** : Do we have all the records that we should? Do we have missing records or not? Are there specific rows, columns, or cells missing?
- **Validity** : We have the records, but they're not valid. ie., they don't conform to a defined schema. A schema is a defined set of rules for data. These rules can be real-world constraints. (e.g. negative height is impossible) and table-specific constraints (e.g. unique key constraints in tables).
- **Accuracy** : Inaccurate data is wrong data that is valid. It adheres to the defined schema, but i is still incorrect. (e.g. A patients weight that is 5 lbs too heavy because the scale was faulty.)
- **Consistency** : Inconsistent data is both valid and accurate, but there are multiple correct ways of referring to the same thing. Consistency, i.e., a standard format, in columns that represent the same data across tables and/or within tables is desired.

Every information here is generated inside the lyft system so it's impossible to get the information to fill unless we hack the lyft backend. We will check the columns and decide it to drop the n/a or not.

***Code***

Let's merge the duplicated columns first!

```python
# We fill the n/a by '' to add the 2 string columns to one!
clean_df.start_time = clean_df.started_at.fillna('') + clean_df.start_time.fillna('')
clean_df.end_time = clean_df.ended_at.fillna('') + clean_df.end_time.fillna('')

# change the '' to n/a again
clean_df.start_time = clean_df.start_time.replace('', np.nan)
clean_df.end_time = clean_df.end_time.replace('', np.nan)

# Remove the unnecessary columns
clean_df.drop(columns=['started_at', 'ended_at'], inplace=True)
```

```python
# We fill the n/a by 0 to add the 2 float64 columns to one!
clean_df.start_station_latitude = clean_df.start_lat.fillna(0) + clean_df.start_station_latitude.fillna(0)
clean_df.start_station_longitude = clean_df.start_lng.fillna(0) + clean_df.start_station_longitude.fillna(0)
clean_df.end_station_latitude = clean_df.end_lat.fillna(0) + clean_df.end_station_latitude.fillna(0)
clean_df.end_station_longitude = clean_df.end_lng.fillna(0) + clean_df.end_station_longitude.fillna(0)

# change the 0 to n/a again
clean_df.start_station_latitude = clean_df.start_station_latitude.replace(0, np.nan)
clean_df.start_station_longitude = clean_df.start_station_longitude.replace(0, np.nan)
clean_df.end_station_latitude = clean_df.end_station_latitude.replace(0, np.nan)
clean_df.end_station_longitude = clean_df.end_station_longitude.replace(0, np.nan)

# Remove the unnecessary columns
clean_df.drop(columns=['start_lat', 'start_lng', 'end_lat', 'end_lng'], inplace=True)
```

```python
# We fill the n/a by '' to add the 2 string columns to one!
clean_df.user_type = clean_df.member_casual.fillna('') + clean_df.user_type.fillna('')

# change the '' to n/a again
clean_df.user_type = clean_df.user_type.replace('', np.nan)

# Remove the unnecessary columns
clean_df.drop(columns=['member_casual'], inplace=True)
```

- **start_station_longitude** - The longitude of the station where the user borrowed the bike
- **end_station_latitude** - The latitude of the station the user returned the bike
- **end_station_longitude** - The longitude of the station where the user returned the bike
    - `float64` dtype.
<br><br>
- **user_type** - Is the user subscriber(member) of service or just casual user(Customer)?
- **rental_access_method** - The method the user used to pay when they borrow the bike (app, clipper)

### Datetime the user mostly start to use the bike

```python
# make a new copy of original
bike_df = df.copy()

# Generate necessary columns
bike_df['year'] = bike_df.start_time.dt.year
bike_df['month'] = bike_df.start_time.dt.month
bike_df['day'] = bike_df.start_time.dt.day
bike_df['hour'] = bike_df.start_time.dt.hour

# leave the necessary columns
bike_df = bike_df[['year', 'month', 'day', 'hour']]

# Remove NaN values
bike_df = bike_df[~(bike_df.year.isna() |
                    bike_df.month.isna() |
                    bike_df.day.isna() |
                    bike_df.hour.isna())]

# Check values
bike_df.sample(5)
```

```python
columns = ['year', 'month', 'day', 'hour']

# figure size
plt.figure(figsize=[15, 40])

# permutation without i exceeding j
# ex) [0, 1], [0, 2], [0, 3], [1, 2], [1, 3], [2, 3]
ax_num = 1
for i in range(4):
    for j in range(i+1, 4):
        # choose the axes
        plt.subplot(6, 1, ax_num)
        ax_num += 1
        
        # Choose what columns will be used for plot
        row = columns[i]
        column = columns[j]
        
        # groupby with row and column 
        bike_table = bike_df.groupby(['{}'.format(row), '{}'.format(column)]).count().reset_index()
        
        # Get the remain column
        k = list(set([0, 1, 2, 3]) ^ set([i, j]))[0]
        count = columns[k]
        
        # pivot the table to be able to apply heatmap
        bike_table = bike_table.pivot(index='{}'.format(row),
                                      columns='{}'.format(column),
                                      values='{}'.format(count))
        
        # plot the heatmap
        if ax_num == 2:
            annot_bool = True
        else:
            annot_bool = False
        sb.heatmap(bike_table, annot=annot_bool, fmt='g', cmap='YlGnBu')
        plt.yticks(rotation=1)
        plt.ylabel('{}'.format(row), rotation=1);
```

### Distance & bike usage time of user's bike travel per start hour

```python
# make a new copy of original
bike_df = df.copy()

# leave the necessary columns
bike_df = bike_df[['start_time',
                   'start_station_longitude',
                   'end_station_longitude',
                   'start_station_latitude',
                   'end_station_latitude',
                   'duration_sec']]

# Remove NaN values
bike_df = bike_df[~(bike_df.start_time.isna() |
                    bike_df.start_station_longitude.isna() |
                    bike_df.end_station_longitude.isna() |
                    bike_df.start_station_latitude.isna() |
                    bike_df.end_station_latitude.isna() |
                    bike_df.duration_sec.isna())]

# Remove the values which the destination is same as the start station
bike_df = bike_df.query('start_station_longitude != end_station_longitude or start_station_latitude != end_station_latitude')

# Check values
bike_df.sample(5)
```

```python
# Change the start_time to hour
bike_df.start_time = bike_df.start_time.dt.hour

# calculate the difference of longitude between start and end
# apply abs
# change it to mile 1 degree of longitude = 69.172 miles
bike_df.start_station_longitude = bike_df.start_station_longitude - bike_df.end_station_longitude
bike_df.start_station_longitude = bike_df.start_station_longitude.map(abs)
bike_df.start_station_longitude = bike_df.start_station_longitude*69.172

# calculate the difference of latitude between start and end
# apply abs
# change it to mile 1 degree of longitude = 69 miles
bike_df.start_station_latitude = bike_df.start_station_latitude - bike_df.end_station_latitude
bike_df.start_station_latitude = bike_df.start_station_latitude.map(abs)
bike_df.start_station_latitude = bike_df.start_station_latitude*69

# Change second to hour
bike_df.duration_sec = bike_df.duration_sec/3600

# rename columns
bike_df.rename(columns={'start_time': 'start_hour',
                        'start_station_longitude': 'mile_lng',
                        'start_station_latitude': 'mile_lat',
                        'duration_sec': 'duration_hour'},
               inplace=True)

# calculate the total distance as mile
# When we consider the complicated roads on the city
# I calculate the distance as manhattan distance
bike_df.mile_lng = bike_df.mile_lng + bike_df.mile_lat

# rename mile_lng
bike_df.rename(columns={'mile_lng': 'distance_mile'}, inplace=True)

# drop the unnecessary columns
bike_df.drop(columns=['end_station_longitude', 'end_station_latitude', 'mile_lat'], inplace=True)
bike_df
```
